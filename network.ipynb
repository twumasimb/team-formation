```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty undirected graph
G = nx.Graph()

# Sample data: list of dictionaries
data = [
    {"id": 1, "name": "Alice", "skills": ["Python", "Java"]},
    {"id": 2, "name": "Bob", "skills": ["Python", "C++"]},
    {"id": 3, "name": "Charlie", "skills": ["Java", "JavaScript"]},
    # ... more dictionaries ...
]

# Add nodes with id as the node name
for person in data:
    G.add_node(person["id"], name=person["name"])

# Add edges based on common skills with weights
for i, person1 in enumerate(data):
    for j, person2 in enumerate(data[i+1:], start=i+1):
        common_skills = set(person1["skills"]) & set(person2["skills"])
        if common_skills:
            weight = len(common_skills)
            G.add_edge(person1["id"], person2["id"], common_skills=common_skills, weight=weight)

# Draw the graph with weighted edges
pos = nx.spring_layout(G)
edge_labels = {(u, v): data["weight"] for u, v, data in G.edges(data=True)}
nx.draw(G, pos, with_labels=True, node_size=1000, font_size=12, font_color="black", font_weight="bold")
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

# Display the graph
plt.show()
```
